Softmax da una ponderación entre 0 y 1 a los scores/logits que entran a esa función. 

```python
import torch
import torch.nn as nn
import numpy as np
```

```python
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
x = np.array([2.0, 1.0, 0.1])
print(f'softmax of X is {softmax(x)}')
```

```python
# Now in Pytorch
x_t = torch.from_numpy(x)
outputs_t = torch.softmax(x_t, dim=0) 
# dim=0 computes along first axis
outputs_t
```

# Cross entropy loss
Measures the performance of a classification model of a model with probability between 0 and 1.

The loss increases as the probabilty diverges from the predicted label

```python
def cross_entropy(actual, predicted):
  loss_ce = 0
  loss_ce -= np.sum(actual * np.log(predicted))
  return loss_ce
```

```python
Y = np.array([1, 0, 0])

pred_good = np.array([0.7, 0.2, 0.1])
pred_bad = np.array([0.1, 0.3, 0.6])

print(f'Low CE on good prediction: {cross_entropy(Y, pred_good)}; High CE on bad prediction: {cross_entropy(Y, pred_bad)}')
```

```python
# In Pytorch
# Cross Entropy Loss applies Softmax + NLLLoss (Negative log likelihood loss)
# Do not implement softmax in last layer
# Y has class labels, not one hot
# Y must have raw scores, not softmax

# Three possible classes
Y = torch.tensor([2, 0, 1])
# number of samples * number of classes
Y_pred_good = torch.tensor([[0.1, 1.0, 2.1], [2.0, 1.0, 0.1], [1.0, 2.0, 0.1]])
Y_pred_bad = torch.tensor([[0.5, 2.0, 0.3], [0.5, 2.0, 0.3], [1.5, 0.0, 0.3]])

loss = nn.CrossEntropyLoss()

print(f'Low CE on good prediction: {loss(Y_pred_good, Y)}; High CE on bad prediction: {loss(Y_pred_bad, Y)}')
```

```python
# to get the predictions
# Get the position of the tensor with the highest value

_, predictions = torch.max(Y_pred_good, 1)
_, predictions_2 = torch.max(Y_pred_bad, 1)
print(predictions, predictions_2)
```
