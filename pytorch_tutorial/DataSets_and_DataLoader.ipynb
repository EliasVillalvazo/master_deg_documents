```python
import torch
import torchvision
from torch.utils.data import Dataset, DataLoader

import numpy as np
import math
```

```python
class WineDataSet(Dataset):
  def __init__(self):
    # data loading
    data = np.loadtxt('./wine.csv', delimiter=',', dtype=np.float32, skiprows=1)
    self.x = torch.from_numpy(data[:, 1:])
    self.y = torch.from_numpy(data[:, [0]])
    self.n_samples = data.shape[0]

  def __getitem__(self, index):
    # Returns a tupple containing the features and the label at <index>
    return self.x[index], self.y[index]
  
  def __len__(self):
    return self.n_samples

dataset = WineDataSet()
# Create a dataloader
dataloader = DataLoader(dataset=dataset, batch_size=4, shuffle=True, num_workers=2)

# dataiter = iter(dataloader)
# data = dataiter.next()
# features, labels = data
# print(features, labels)
```

```python
# create a dummy training loop
num_epochs = 2
total_samples = len(dataset)
batch_size = 4 
num_iters = math.ceil(total_samples / batch_size)

print(total_samples, num_iters)
```

```python
for epoch in range(num_epochs):
  for i, (inputs, labels) in enumerate(dataloader):
    # Perform forward and backward pass
    if (i+1) % 5 == 0:
      print(f'epoch: {epoch+1}/{num_epochs}; step {i+1}/{num_iters}; inputs: {inputs.shape}')
```
